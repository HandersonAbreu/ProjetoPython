```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Limpando o df
df.drop(['codigo_ocorrencia1', 'codigo_ocorrencia2',
       'codigo_ocorrencia3', 'codigo_ocorrencia4',
       'ocorrencia_latitude', 'ocorrencia_longitude',
       'ocorrencia_aerodromo','ocorrencia_hora',
       'investigacao_aeronave_liberada','divulgacao_relatorio_numero',
       'divulgacao_relatorio_publicado', 'divulgacao_dia_publicacao',
       'total_recomendacoes', 'total_aeronaves_envolvidas',
       'ocorrencia_saida_pista'], axis= 1, inplace=True)
```

```python
#Como ficou
df.columns
```

```python
#Renomeando as colunas
df.columns = ['Código da ocorrência', 'Classificação', 'Cidade', 'UF', 'País',
              'Dia da ocorrência', 'Status da investigação']
#Adcionando coluna para contagem
df['Contagem'] = 1
```

```python
df.head()
```

```python
classificacao = df.groupby('Classificação')['Contagem'].sum().plot(kind='bar', figsize=(12,7), color='blue', alpha=.6)

for p in classificacao.patches:
    b=p.get_bbox()
    classificacao.annotate((round(b.y1 + b.y0)), ((b.x0 + b.x1)/2 - 0.03, b.y1 + 15))

plt.xticks(rotation=0)
plt.xlabel('Classificação da Ocorrência', weight='bold', fontsize=12)
plt.ylabel('Quantidade', weight='bold', fontsize=12)
plt.title('Quantidade de ocorrências por classificação\n Total: ' + str(df['Contagem'].sum()) + ' de ocorrências', weight='bold', fontsize=14)
plt.show()
```

```python
plotagem_data = df.copy()
plotagem_data['Dia da ocorrência'] = plotagem_data['Dia da ocorrência'].apply(lambda x: x.replace(x,x[6:10]))
plotagem_data.groupby('Dia da ocorrência')['Contagem'].sum().plot(color='r', figsize=(14,5), grid=True)

plt.title('Quantidade de ocorrência nos últimos anos \n Total: ' + str(df['Contagem'].sum()) + ' ocorrências', weight='bold', fontsize=14)
plt.xlabel('Ano de referência', weight='bold', fontsize=12)
plt.ylabel('Qauntidade de ocorrências', weight='bold', fontsize=12)
plt.show()
```

```python
estado =  df.groupby('UF')['Contagem'].sum().plot(kind='bar', figsize=(11,5))

plt.xlabel('UF')
plt.ylabel('Quantidade de ocorrência', weight='bold', fontsize=12)
plt.title('Quantidade de ocorrência por Estado\nTotal: ' + str(df['Contagem'].sum()) + ' ocorrências', weight='bold', fontsize=14)
plt.xticks(rotation=0)
plt.show()
```
