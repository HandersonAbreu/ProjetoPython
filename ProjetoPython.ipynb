```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

#Ocorrências

```python
df = pd.read_csv('/content/dados/ocorrencia.csv', delimiter=';', on_bad_lines='skip')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.columns
```

```python
#Limpando o df
df.drop(['codigo_ocorrencia1', 'codigo_ocorrencia2',
       'codigo_ocorrencia3', 'codigo_ocorrencia4',
       'ocorrencia_latitude', 'ocorrencia_longitude',
       'ocorrencia_aerodromo','ocorrencia_hora',
       'investigacao_aeronave_liberada','divulgacao_relatorio_numero',
       'divulgacao_relatorio_publicado', 'divulgacao_dia_publicacao',
       'total_recomendacoes', 'total_aeronaves_envolvidas',
       'ocorrencia_saida_pista'], axis= 1, inplace=True)
```

```python
#Como ficou
df.columns
```

```python
#Renomeando as colunas
df.columns = ['Código da ocorrência', 'Classificação', 'Cidade', 'UF', 'País',
              'Dia da ocorrência', 'Status da investigação']
#Adcionando coluna para contagem
df['Contagem'] = 1
```

```python
df.head()
```

```python
classificacao = df.groupby('Classificação')['Contagem'].sum().plot(kind='bar', figsize=(12,7), color='blue', alpha=.6)

for p in classificacao.patches:
    b=p.get_bbox()
    classificacao.annotate((round(b.y1 + b.y0)), ((b.x0 + b.x1)/2 - 0.03, b.y1 + 15))

plt.xticks(rotation=0)
plt.xlabel('Classificação da Ocorrência', weight='bold', fontsize=12)
plt.ylabel('Quantidade', weight='bold', fontsize=12)
plt.title('Quantidade de ocorrências por classificação\n Total: ' + str(df['Contagem'].sum()) + ' de ocorrências', weight='bold', fontsize=14)
plt.show()
```

```python
plotagem_data = df.copy()
plotagem_data['Dia da ocorrência'] = plotagem_data['Dia da ocorrência'].apply(lambda x: x.replace(x,x[6:10]))
plotagem_data.groupby('Dia da ocorrência')['Contagem'].sum().plot(color='r', figsize=(14,5), grid=True)

plt.title('Quantidade de ocorrência nos últimos anos \n Total: ' + str(df['Contagem'].sum()) + ' ocorrências', weight='bold', fontsize=14)
plt.xlabel('Ano de referência', weight='bold', fontsize=12)
plt.ylabel('Qauntidade de ocorrências', weight='bold', fontsize=12)
plt.show()
```

```python
estado =  df.groupby('UF')['Contagem'].sum().plot(kind='bar', figsize=(14,7))

plt.xlabel('UF')
plt.ylabel('Quantidade de ocorrência', weight='bold', fontsize=12)
plt.title('Quantidade de ocorrência por Estado\nTotal: ' + str(df['Contagem'].sum()) + ' ocorrências', weight='bold', fontsize=14)
plt.xticks(rotation=0)
plt.show()
```

```python
fator_contribuinte.drop(['ocorrencia_tipo_categoria', 'taxonomia_tipo_icao'], axis= 1, inplace=True)
```

#Aéronaves

```python
aeronave = pd.read_csv('/content/dados/aeronave.csv', delimiter=';', on_bad_lines='skip')
```

```python
aeronave.head()
```

```python
aeronave.dtypes
```

```python
aeronave.columns
```

```python
aeronave.drop(['aeronave_matricula', 'aeronave_operador_categoria', 'aeronave_tipo_icao', 'aeronave_pmd',
               'aeronave_pmd_categoria', 'aeronave_pais_fabricante', 'aeronave_registro_categoria',
               'aeronave_registro_segmento', 'aeronave_voo_origem', 'aeronave_voo_destino', 'aeronave_fase_operacao'],
              axis=1, inplace=True)
```

```python
aeronave['Contagem'] = 1
aeronave.head()
```

```python
aeronave.columns
```

```python
aeronave.columns = ['Código da ocorrência', 'Tipo', 'Fabricante', 'Modelo', 'Motor',
                    'Quantidade de Motor', 'Assentos', 'Ano de fabricação', 'País de registro',
                    'Tipo de Operação', 'Nível de dano', 'Fatalidades', 'Contagem']
aeronave = aeronave.apply(lambda x: x.replace('***','Não identificado'))
```

```python
aeronave.head()
```

```python
#Função para converter NaN em 0
def convert_nan(dataframe, column):
    for i in column:
        dataframe[i].fillna(0, inplace = True)

convert_nan(aeronave,['Ano de fabricação','Assentos', 'Fatalidades'])
```

```python
def convert_dtype(dataframe, column, dtype):
    for i in column:
        dataframe[i] = dataframe[i].astype(dtype)

convert_dtype(aeronave,['Ano de fabricação','Assentos'],int)
```

```python
aeronave.dtypes
```

```python
aeronave.sample(10)
```

```python
motor = aeronave.groupby('Motor')['Contagem'].sum().sort_values().plot(kind='bar', figsize=(14,5))

for p in motor.patches:
    b=p.get_bbox()
    motor.annotate((round(b.y1 + b.y0)), ((b.x0 + b.x1)/2 - 0.1, b.y1 + 15))

plt.title('Quantidade de ocorrências\n por tipo de motor\nTotal: '+str(aeronave['Contagem'].sum())+' de ocorrências', weight='bold', fontsize=14)
plt.ylabel('Quantidade de ocorrências', weight='bold', fontsize=12)
plt.xlabel('Tipo de motor', weight='bold', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
tipo = aeronave.groupby('Tipo')['Contagem'].sum().sort_values().plot(kind='bar', figsize=(14,7))

for p in tipo.patches:
    b=p.get_bbox()
    tipo.annotate((round(b.y1 + b.y0)), ((b.x0 + b.x1)/2 - 0.1, b.y1 + 15))

plt.title('Quantidade de ocorrências\n por tipo de aéronave\n Total: '+str(aeronave['Contagem'].sum())+' de ocorrências', weight='bold', fontsize=14)
plt.ylabel('Quantidade de ocorrências', weight='bold', fontsize=12)
plt.xlabel('Tipo de aéronave', weight='bold', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
nivel = aeronave.groupby('Nível de dano')['Contagem'].sum().sort_values().plot(kind='bar', figsize=(10,5))

for p in nivel.patches:
    b=p.get_bbox()
    nivel.annotate((round(b.y1 + b.y0)), ((b.x0 + b.x1)/2 - 0.1, b.y1 + 15))

plt.title('Quantidade de ocorrências\n por nível de dano\nTotal: '+str(aeronave['Contagem'].sum())+' de ocorrências', weight='bold', fontsize=14)
plt.ylabel('Quantidade de ocorrências', weight='bold', fontsize=12)
plt.xlabel('Nível de dano', weight='bold', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
fatalidades = aeronave[aeronave['Fatalidades'] > 0]
fatalidade = fatalidades.groupby('Fatalidades')['Contagem'].sum().plot(kind='bar', figsize=(10,6))

for p in fatalidade.patches:
    b=p.get_bbox()
    fatalidade.annotate((round(b.y1 + b.y0)), ((b.x0 + b.x1)/2 - 0.1, b.y1 + 2))

plt.title('Quantidade de ocorrências\n por total de fatalidades\nTotal: '+str(fatalidades['Contagem'].sum())+' de ocorrências', weight='bold', fontsize=14)
plt.ylabel('Quantidade de fatalidades', weight='bold', fontsize=12)
plt.xlabel('Fatalidades', weight='bold', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```
